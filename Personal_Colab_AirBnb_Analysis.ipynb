## ***Data Wrangling***

```python
# making acopy of dataframe.
airbnb = df.copy() 
```

```python
# filling na values 
airbnb.fillna({'reviews_per_month':0},inplace=True)
airbnb.fillna({'host_name':'Not Mentioned'}, inplace = True)
```

#Replacing all NaN values in review_per_month_ with 0 and host_name with Not Mentioned

```python
#dropping columns that are not significant for our future data exploration 
airbnb.drop(['id','last_review','name'],axis=1,inplace=True)
```

#We have 3 columns that are not significant for our future data exploration. Now, we drop unnecessary columns such as id, name, last_review.

```python
# renaming some column name for better understanding
airbnb.rename(columns= {'neighbourhood_group': 'Location', 'neighbourhood': 'Area'},inplace=True)
```

Lets see the boxplot of this price column to have a feel of the presence of outliers. Dont worry, Ill be handling these outlier values!

```python
sns.boxenplot(airbnb['price'],orient='horizontal')
```

```python
airbnb['price'].describe()
```

#**Removing outliers from price column**

```python
# Select the column where you want to remove outliers
column_name = 'price'

# Calculate the IQR and define the threshold
q1 = airbnb[column_name].quantile(0.25)
q3 = airbnb[column_name].quantile(0.75)
iqr = q3 - q1
threshold = 1.5

# Remove the outliers
filtered_airbnb = airbnb[(airbnb[column_name] >= q1 - threshold*iqr) & (airbnb[column_name] <= q3 + threshold*iqr)]
```

```python
#Now, lets check the boxen plot once again as by now I have removed the outliers.
sns.boxenplot(filtered_airbnb['price'],orient='horizontal')
```


# **Let's start with data analysis on Airbnb dataset**

```python
# Chart-1: Plot of distribution of types of rooms booked.

#set figure size
plt.figure(figsize=(9,6))

#set up plot title
plt.title('Type of Rooms Booked', {'fontsize': 18, 'fontweight': 20})

#plot count plot
sns.countplot(x='room_type',data=airbnb)

#add up labels of x and y
plt.xlabel("Room Type", {'fontsize': 12, 'fontweight': 15})
plt.ylabel("No of Bookings", {'fontsize': 12, 'fontweight': 15})
```

## **which region is the most expensive region and what are effect of these on bookings?**

```python
 #Plot between Map of Property Type Distribution vs Map of Price Distribution

# Create a figure with two subplots
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(20, 6))

# Plot first subplot
sns.scatterplot(ax=axs[0], x='longitude', y='latitude', hue='Location', data=airbnb)

axs[0].set_title("Map of Room Type Distribution", {'fontsize': 12, 'fontweight': 15})

# Set label
axs[0].set_xlabel('Longitude', {'fontsize': 10, 'fontweight': 12})
axs[0].set_ylabel('Latitude', {'fontsize': 10, 'fontweight': 12})



# Plot second subplot
sns.scatterplot(ax=axs[1], data=airbnb[airbnb.price<500], x="longitude", y="latitude", hue="price",palette='plasma')
axs[1].legend(title="Price", bbox_to_anchor=(1.05, 1), loc=2)
axs[1].set_title("Map of Price Distribution", {'fontsize': 12, 'fontweight': 15})

# Set title
axs[0].set_xlabel('Longitude', {'fontsize': 10, 'fontweight': 12})
axs[0].set_ylabel('Latitude', {'fontsize': 10, 'fontweight': 12})
# Display the figure
plt.show()
```
