## ***Data Wrangling***

```python
# making acopy of dataframe.
airbnb = df.copy() 
```

```python
# filling na values 
airbnb.fillna({'reviews_per_month':0},inplace=True)
airbnb.fillna({'host_name':'Not Mentioned'}, inplace = True)
```

#Replacing all NaN values in âreview_per_month_ with 0 and host_name with Not Mentioned

```python
#dropping columns that are not significant for our future data exploration 
airbnb.drop(['id','last_review','name'],axis=1,inplace=True)
```

#We have 3 columns that are not significant for our future data exploration. Now, we drop unnecessary columns such as id, name, last_review.






```python
# Lets add a new column named Price range

def get_price_range(pr):
  if pr>= 0 and pr<200:
    price_range = "0-200"
  elif pr>= 200 and pr<400:
    price_range = "200-400"
  elif pr>= 400 and pr<600:
    price_range = "400-600"
  elif pr>= 600 and pr<800:
    price_range = "600-800"
  elif pr>= 800 and pr<1000:
    price_range = "800-1000"
  else:
    price_range = ">1000"
  return price_range
```

```python
airbnb['price_range'] = airbnb.apply(lambda x: get_price_range(x['price']), axis= 1)
```

```python
# renaming some column name for better understanding
airbnb.rename(columns= {'neighbourhood_group': 'Location', 'neighbourhood': 'Area'},inplace=True)
```

#**Removing outliers from price column**

```python
# Select the column where you want to remove outliers
column_name = 'price'

# Calculate the IQR and define the threshold
q1 = airbnb[column_name].quantile(0.25)
q3 = airbnb[column_name].quantile(0.75)
iqr = q3 - q1
threshold = 1.5

# Remove the outliers
filtered_airbnb = airbnb[(airbnb[column_name] >= q1 - threshold*iqr) & (airbnb[column_name] <= q3 + threshold*iqr)]
```

```python
#Now, letâs check the box plot once again as by now I have removed the outliers.
sns.boxplot(filtered_airbnb['price'],orient='horizontal')
```

##**4 .**  **which region is the most expensive region and what are effect of these on bookings?**

```python
# Chart - 4: Plot between Map of Property Type Distribution vs Map of Price Distribution

# Create a figure with two subplots
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(20, 6))

# Plot first subplot
sns.scatterplot(ax=axs[0], x='longitude', y='latitude', hue='Location', data=airbnb)

axs[0].set_title("Map of Room Type Distribution", {'fontsize': 12, 'fontweight': 15})

# Set label
axs[0].set_xlabel('Longitude', {'fontsize': 10, 'fontweight': 12})
axs[0].set_ylabel('Latitude', {'fontsize': 10, 'fontweight': 12})



# Plot second subplot
sns.scatterplot(ax=axs[1], data=airbnb, x="longitude", y="latitude", hue="price_range")
axs[1].legend(title="Price", bbox_to_anchor=(1.05, 1), loc=2)
axs[1].set_title("Map of Price Distribution", {'fontsize': 12, 'fontweight': 15})

# Set title
axs[0].set_xlabel('Longitude', {'fontsize': 10, 'fontweight': 12})
axs[0].set_ylabel('Latitude', {'fontsize': 10, 'fontweight': 12})
# Display the figure
plt.show()
```


#The region that stands out as the most expensive is **Manhattan**. The high cost of accommodations in this region significantly impacts the booking trends. Despite its expensiveness, **Manhattan** remains a popular choice for bookings, and also from the previous visualisation we find that the **Entire Home/Apartment** is listed mostly near **Manhattan** means most guest prefered to stay in indipendent property.
