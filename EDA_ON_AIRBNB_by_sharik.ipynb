## 3. ***Data Wrangling***

```python
# making acopy of dataframe.
airbnb = df.copy() 
```

```python
# filling na values 
airbnb.fillna({'reviews_per_month':0},inplace=True)
airbnb.fillna({'host_name':'Not Mentioned'}, inplace = True)
```

```python
#dropping columns that are not significant for our future data exploration and predictions
airbnb.drop(['id','last_review','name'],axis=1,inplace=True)
```

```python
# Lets add a new column named Price range

def get_price_range(pr):
  if pr>= 0 and pr<200:
    price_range = "0-200"
  elif pr>= 200 and pr<400:
    price_range = "200-400"
  elif pr>= 400 and pr<600:
    price_range = "400-600"
  elif pr>= 600 and pr<800:
    price_range = "600-800"
  elif pr>= 800 and pr<1000:
    price_range = "800-1000"
  else:
    price_range = ">1000"
  return price_range
```

```python
airbnb['price_range'] = airbnb.apply(lambda x: get_price_range(x['price']), axis= 1)
```

```python
# renaming some column name for better understanding

# create a dictionary of old column names as keys and new column names as values
airbnb.rename(columns= {'neighbourhood_group': 'Location', 'neighbourhood': 'Area'},inplace=True)




```
